```python
import scipy.io
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
folder_path = "12k 0 load data"
file_paths = []
file_names = []

# The search party looks for .mat files
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.mat'):
            file_paths.append(os.path.join(root, file))
            file_names.append(file.replace('.mat', '')) # Clean up the name for our table

print(f"Cell 2 Complete: Found {len(file_paths)} files.")
print(f"Here is the first file path to verify: {file_paths[0]}")
```

```python
# Open just the first file
test_mat = scipy.io.loadmat(file_paths[0])

print(f"Looking inside: {file_names[0]}")
print("Here are the exact dictionary keys inside this file:")
print(test_mat.keys())
```

```python
# Create an empty dictionary to hold our columns
data_dictionary = {}

for i in range(len(file_paths)):
    path = file_paths[i]
    name = file_names[i]
    mat_data = scipy.io.loadmat(path)
    
    # Safely hunt for the Drive End data
    for key in mat_data.keys():
        if 'DE_time' in key:
            # Flatten the data and take the first 120,000 points so all columns are equal length
            signal = mat_data[key].flatten()[:120000] 
            data_dictionary[name] = signal
            break # Stop looking once we find it


df = pd.DataFrame(data_dictionary)

print("Cell 4 Complete: DataFrame successfully created!")
display(df.head())
```
