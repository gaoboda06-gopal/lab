Variational Mode Decomposition (VMD)

```python
import numpy as np 
import time
import vmdpy as vmd
```

```python
fs=5000
x=np.random.rand(5000)

alpha=2000
tau=0
K=8
DC = 0
init = 1
tol = 1e-7

start = time.time()
u,u_hat,omega = vmd.VMD(x, alpha, tau, K, DC, init, tol)
elapsed = time.time() - start
print(f"VMD runtime for K={K}, N=5000: {elapsed*1000:.1f} ms")
```
